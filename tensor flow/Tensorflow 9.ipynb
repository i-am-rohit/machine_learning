```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow.compat.v1 as tf
```

```python
x_data = house_data.drop(["medianHouseValue"],axis=1)
y_data = house_data["medianHouseValue"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = pd.DataFrame(data=scaler.transform(X_train),columns=X_train.columns,index=X_train.index)
X_train.head()
```

```python
X_test = pd.DataFrame(data=scaler.transform(X_test),columns=X_test.columns,index=X_test.index)
X_test.head()
```

```python
house_data.columns
```

```python
age = tf.feature_column.numeric_column("housingMedianAge")
rooms = tf.feature_column.numeric_column("totalRooms")
bedrooms = tf.feature_column.numeric_column("totalBedrooms")
pop = tf.feature_column.numeric_column("population")
households = tf.feature_column.numeric_column("households")
income = tf.feature_column.numeric_column("medianIncome")
```

```python
feat_cols = [age,rooms,bedrooms,pop,households,income]
```

```python
input_func = tf.estimator.inputs.pandas_input_fn(X_train,y_train,batch_size=10,num_epochs=1000,shuffle=True)
```

```python
model = tf.estimator.DNNRegressor(hidden_units=[10,10,10],feature_columns=feat_cols)
```

```python
model.train(input_func,steps=20000)
```

```python
eval_input_func = tf.estimator.inputs.pandas_input_fn(X_test,y_test,batch_size=100,num_epochs=1,shuffle=False)
```

```python
results = model.evaluate(eval_input_func)
```

```python
results 
```

```python
predict_input_func = tf.estimator.inputs.pandas_input_fn(X_test,batch_size=10,num_epochs=1,shuffle=False)
```

```python
pred_generator = model.predict(predict_input_func)
```

```python
predictions = list(pred_generator)
```
